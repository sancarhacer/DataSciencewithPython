 Understanding Box Plots
 Box plots (also called whisker plots) are used to visualize the distribution of a dataset.
 They provide insights into:
 - The median (Q2, 50th percentile)
 - The interquartile range (IQR) - spread between Q1 (25th percentile) and Q3 (75th percentile)
 - The minimum and maximum values within 1.5 * IQR (whiskers)
 - Outliers (values outside the whiskers)

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.array([5, 7, 9, 15, 20, 22, 25, 30, 32, 35, 37, 40, 50, 55, 60, 100])

plt.figure(figsize=(6, 5))
sns.boxplot(y=data)
plt.title("Box Plot of Simple Numeric Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

### Interpretation ###
 1. The box represents the interquartile range (IQR), which contains 50% of the data.
 2. The line inside the box is the median (50th percentile).
 3. The whiskers extend to 1.5 times the IQR.
 4. Any points beyond the whiskers are considered outliers.
 5. Higher class passengers generally paid higher fares, but there are extreme outliers (very high fares).

```python
df = sns.load_dataset("titanic")
df
```

```python
plt.figure(figsize=(15, 5))

# plt.subplot(1, 2, 1): Figure'ü 1 satır ve 2 sütuna böl ve şu an 1. pozisyonda (sol) çalış.
plt.subplot(1, 2, 1)

# sns.boxplot(): Seaborn kütüphanesini kullanarak kutu grafiği çiz.
# x="class": X ekseninde "class" (yolcu sınıfı: 1., 2., 3.) değişkenini kullan.
# y="age": Y ekseninde "age" (yaş) değişkenini kullan.
# data=df: Kullanılacak veri kaynağını belirt.
sns.boxplot(x="class", y="age", data=df)

plt.title("Box Plot of Age by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Arka plana ızgara çizgileri ekle (veri noktalarını okumayı kolaylaştırır).
plt.grid(True)


# plt.subplot(1, 2, 2): Figure'de 1 satır, 2 sütun düzeninde 2. pozisyonda (sağ) çalış.
plt.subplot(1, 2, 2)

# Box plot for Fare grouped by Class: Ücret (Fare) dağılımını yolcu sınıflarına göre çiz.
sns.boxplot(x="class", y="fare", data=df)


plt.title("Box Plot of Fare by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

# plt.yscale("log"): Y eksenini logaritmik ölçeğe dönüştür.
# Ücret (fare) dağılımı genellikle çarpıktır; bu, yüksek ücretlerin (outlier'lar) daha iyi görünmesini sağlar.
plt.yscale("log")

# Arka plana ızgara çizgileri ekle.
plt.grid(True)

# ------------------------- GENEL DÜZENLEME VE GÖSTERİM -------------------------

# plt.tight_layout(): Alt grafiklerin (subplot'ların) birbirine ve Figure kenarlarına
# çakışmasını engellemek için aralıkları otomatik olarak ayarlar.
plt.tight_layout()

# Oluşturulan Figure'ü ekranda göster.
plt.show()
```

 Higher class passengers generally paid higher fares, but there are extreme outliers (very high fares).
