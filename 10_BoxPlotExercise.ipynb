 Understanding Box Plots
 Box plots (also called whisker plots) are used to visualize the distribution of a dataset.
 They provide insights into:
 - The median (Q2, 50th percentile)
 - The interquartile range (IQR) - spread between Q1 (25th percentile) and Q3 (75th percentile)
 - The minimum and maximum values within 1.5 * IQR (whiskers)
 - Outliers (values outside the whiskers)

```python
import numpy as np
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt
```

```python
data = np.array([5, 7, 9, 15, 20, 22, 25, 30, 32, 35, 37, 40, 50, 55, 60, 100])

plt.figure(figsize=(6, 5))
sns.boxplot(y=data)
plt.title("Box Plot of Simple Numeric Data")
plt.ylabel("Value")
plt.grid(True)
plt.show()
```

### Interpretation ###
 1. The box represents the interquartile range (IQR), which contains 50% of the data.
 2. The line inside the box is the median (50th percentile).
 3. The whiskers extend to 1.5 times the IQR.
 4. Any points beyond the whiskers are considered outliers.
 5. Higher class passengers generally paid higher fares, but there are extreme outliers (very high fares).

```python
df = sns.load_dataset("titanic")
df
```

```python
plt.figure(figsize=(15, 5))

# plt.subplot(1, 2, 1): Figure' 1 satr ve 2 stuna bl ve u an 1. pozisyonda (sol) al.
plt.subplot(1, 2, 1)

# sns.boxplot(): Seaborn ktphanesini kullanarak kutu grafii iz.
# x="class": X ekseninde "class" (yolcu snf: 1., 2., 3.) deikenini kullan.
# y="age": Y ekseninde "age" (ya) deikenini kullan.
# data=df: Kullanlacak veri kaynan belirt.
sns.boxplot(x="class", y="age", data=df)

plt.title("Box Plot of Age by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Age")

# Arka plana zgara izgileri ekle (veri noktalarn okumay kolaylatrr).
plt.grid(True)


# plt.subplot(1, 2, 2): Figure'de 1 satr, 2 stun dzeninde 2. pozisyonda (sa) al.
plt.subplot(1, 2, 2)

# Box plot for Fare grouped by Class: cret (Fare) dalmn yolcu snflarna gre iz.
sns.boxplot(x="class", y="fare", data=df)


plt.title("Box Plot of Fare by Class")
plt.xlabel("Passenger Class")
plt.ylabel("Fare")

# plt.yscale("log"): Y eksenini logaritmik lee dntr.
# cret (fare) dalm genellikle arpktr; bu, yksek cretlerin (outlier'lar) daha iyi grnmesini salar.
plt.yscale("log")

# Arka plana zgara izgileri ekle.
plt.grid(True)

# ------------------------- GENEL DZENLEME VE GSTERM -------------------------

# plt.tight_layout(): Alt grafiklerin (subplot'larn) birbirine ve Figure kenarlarna
# akmasn engellemek iin aralklar otomatik olarak ayarlar.
plt.tight_layout()

# Oluturulan Figure' ekranda gster.
plt.show()
```
