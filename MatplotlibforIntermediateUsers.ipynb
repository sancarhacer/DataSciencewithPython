## Matplotlibforintermediateusers

Figure	Tüm grafiği kapsayan en üst düzey kap (pencere/tuval).

Axes	Asıl verinin çizildiği, X ve Y eksenlerini içeren alan (alt grafik).

Spines	Axes alanını çevreleyen çizgiler (çerçeve).

Ticks	Eksenler üzerindeki küçük işaretler (ana ve ikincil).

Labels	Ticks'lerin yanındaki sayısal veya metinsel değerler.

Title	Axes veya Figure'ün başlığı.

Legend	Grafikteki çizgilerin/noktaların ne anlama geldiğini gösteren anahtar.

```python
import matplotlib.pyplot as plt
import numpy as np

# 3x3'lük bir alt grafik matrisi oluştur
fig, axs = plt.subplots(3, 3, figsize=(6, 6))

# Sol üstteki alt grafiğin (0. satır, 0. sütun) arka plan rengini maviye yakın bir tona ayarla
axs[0, 0].set_facecolor("#ddddff")

# Sağ alttaki alt grafiğin (2. satır, 2. sütun) arka plan rengini sarıya yakın bir tona ayarla
axs[2, 2].set_facecolor("#ffffdd")

# Başlık ekleyelim
axs[0, 0].set_title("Mavi Arka Plan", fontsize=8)
axs[2, 2].set_title("Sarı Arka Plan", fontsize=8)

plt.tight_layout()
plt.show()
```

Spines, Axes'i çevreleyen çizgilerdir. Genellikle üst ve sağ kenarları kaldırarak grafiğe daha minimalist bir görünüm kazandırmak istenir.

```python
fig, ax = plt.subplots(figsize=(4, 3))
X = np.linspace(0, 10, 100)
ax.plot(X, np.sin(X))

# 'top' (üst) çerçeveyi gizle (rengini 'None' yaparak)
# ax.spines["top"].set_color("None")
ax.spines["top"].set_color("red")

# 'right' (sağ) çerçeveyi gizle
# ax.spines["right"].set_color("None")
ax.spines["right"].set_color("blue")

ax.set_title("Üst kırmızı ve Sağ  mavi Çerçeve oldu")
plt.show()
```

Ticks ve Etiketler (Ticks & Labels)

Bu, eksen işaretlerini ve etiketlerini çok ince detayda kontrol etmeyi sağlar. Özellikle bilimsel grafiklerde, ikincil (minor) işaretlerin görünmesi önemlidir.

```python
# Ticker modülünden gerekli sınıfları içe aktar
from matplotlib.ticker import MultipleLocator as ML
from matplotlib.ticker import ScalarFormatter as SF

fig, ax = plt.subplots(figsize=(6, 4))
X = np.linspace(0, 5, 100)
ax.plot(X, np.sin(X) + 1)
ax.set_ylim(0, 2)

# X ekseninde ana işaretleyicileri 1 birim aralıkla ayarla (varsayılan)
ax.xaxis.set_major_locator(ML(1.0))

# X ekseninde ikincil (minor) işaretleyicileri 0.2 birim aralıkla ayarla
ax.xaxis.set_minor_locator(ML(0.2))

# İkincil işaretleyiciler için formatı ayarla (SF() standart format demektir)
ax.xaxis.set_minor_formatter(SF())

# X ekseni işaretleyici parametrelerini ayarla
ax.tick_params(
    axis='x',              # X eksenini hedef al
    which='minor',         # İkincil (minor) işaretleyicileri hedef al
    rotation=45,           # Etiketleri 45 derece döndür
    labelsize=8,           # Etiket yazı boyutunu küçült
    color='red'            # İşaret çizgisi rengini kırmızı yap
)

ax.set_title("Özelleştirilmiş Major ve Minor Ticks")
plt.show()
```

Lines & Markers (Çizgiler ve İşaretleyiciler)

Bu kısayol notasyonu, plot fonksiyonunda çizgi stilini, işaretleyici tipini ve rengi tek bir dize ile hızlıca ayarlamanızı sağlar.

Format String: "C1o:"

*   C1: Matplotlib'in varsayılan renk döngüsündeki ikinci rengi (genellikle turuncu).

*   o: Daire şeklinde işaretleyici (marker).
*   :: Noktalı çizgi (linestyle).







```python
fig, ax = plt.subplots(figsize=(6, 3))
X = np.linspace(0.1, 10 * np.pi, 1000)
Y = np.sin(X)

# C1 rengi (turuncu), 'o' işaretleyicisi, ':' noktalı çizgi stili kullan
# markevery=50: Her 50 veri noktasında bir işaretleyici çiz
# mec="1.0": Marker Edge Color (işaretleyici kenar rengi) beyaz (1.0 = beyaz) yap
ax.plot(X, Y, "C1o:", markevery=50, mec="1.0")

ax.set_title("Kısayol Notasyonu: 'C1o:'")
plt.show()
```

Ölçekler ve Projeksiyonlar (Scales & Projections)

Normal (lineer) eksenler yerine, logaritmik veya diğer ölçekleri kullanabilirsiniz. Bu, geniş aralıktaki verileri tek bir grafikte göstermek için idealdir.

```python
fig, ax = plt.subplots(figsize=(6, 3))
X = np.logspace(0, 3, 100) # 10^0 (1) ile 10^3 (1000) arasında veri
Y = np.sqrt(X) # Y verisi

# X ekseni ölçeğini logaritmik olarak ayarla
ax.set_xscale("log")

# Çizgi grafiğini çiz
ax.plot(X, Y, "C0o-", markevery=10, mec="1.0")

ax.set_title("Logaritmik X Ekseni Ölçeği (ax.set_xscale('log'))")
plt.show()
```

Açıklamalar ve Düzenlemeler (Legend, Text & Annotation)

Grafiğinizi profesyonel hale getirmek için metinler, dolgular ve açıklamalar ekleyebilirsiniz.

a. Legend (Lejant) Özelleştirme
ax.legend() fonksiyonundaki parametreler, lejantın konumunu, sütun sayısını ve düzenini ince ayar yapmanızı sağlar.

```python
fig, ax = plt.subplots(figsize=(6, 4))
X = np.linspace(0, 10, 100)

# Çizgileri etiketleyerek çiz
ax.plot(X, np.sin(X), "C0", label="Sinüs Sinyali")
ax.plot(X, np.cos(X), "C1", label="Kosinüs Sinyali")

# Lejantı Axes'in üst dışına yerleştirme ve ayarlama
ax.legend(
    bbox_to_anchor=(0, 1.02, 1, 0.102), # (x, y, genişlik, yükseklik) Axes'e göre konum
    ncol=2,                            # 2 sütun kullan (etiketleri yan yana dizer)
    mode="expand",                     # Lejantı verilen genişliğe (1) yay
    loc="lower left",                  # Konumlandırma için referans nokta
    borderaxespad=0.                    # Axes ile lejant arasındaki boşluğu ayarla
)

ax.set_title("Özelleştirilmiş Lejant Konumu")
plt.show()
```

Annotation (Açıklama Ekleme)

Belirli bir veri noktasını ok veya metin ile işaretlemek için kullanılır.

```python
fig, ax = plt.subplots(figsize=(6, 4))
X = np.linspace(0, 10*np.pi, 1000)
Y = np.sin(X)
ax.plot(X, Y)

# Açıklama eklenecek veri noktasını bul
idx = 250
x_point, y_point = X[idx], Y[idx]

# ax.annotate(metin, veri_noktası, metin_konumu)
ax.annotate(
    "Maksimum",                                  # Metin
    (x_point, y_point),                          # Açıklanan veri noktası konumu
    (x_point, y_point + 0.5),                    # Metnin çizileceği konum
    ha="center",                                 # Yatay hizalama: ortala
    arrowprops={                                 # Ok özellikleri sözlüğü
        "arrowstyle": "->",                      # Ok stili: basit ok
        "color": "C1",                           # Ok rengi: turuncu
        "lw": 2                                  # Çizgi kalınlığı
    }
)
ax.set_title("Veri Noktasına Ok ile Açıklama")
plt.show()
```

Çözünürlük ve Boyut (Size & DPI)

Grafiğin baskı veya yayın kalitesi, doğru boyut ve çözünürlük ayarına (DPI - Dots Per Inch) bağlıdır.

```python
# Not: Yayın standartlarına (örneğin bir dergi sütunu) uyacak şekilde
# figür genişliği genellikle cm veya inç cinsinden belirlenir.
# Örnek hesaplama: 8cm genişlik, yaklaşık 3.15 inç'e eşittir.
# fig=plt.figure() yerine fig, ax döndüren plt.subplots() kullanmak önerilir.

# fig, ax = plt.subplots(): Bir Figure (fig) ve tek bir Axes (ax) oluştur.
# figsize=(3.15, 3.15): Figure'ün genişliğini ve yüksekliğini 3.15 inç olarak ayarla (kare).
fig, ax = plt.subplots(figsize=(3.15, 3.15))

# Grafikte gösterilecek rastgele bir veri seti oluştur.
# np.arange(5): [0, 1, 2, 3, 4] X değerleri.
# np.random.rand(5): 0 ile 1 arasında 5 adet rastgele Y değeri.
ax.plot(np.arange(5), np.random.rand(5))

# Axes'e bir başlık ekle.
ax.set_title("Yayın Kalitesi İçin Boyut")

# ----------------- KAYDETME İŞLEMİ -----------------

# fig.savefig("figure_yuksek_kalite.pdf", dpi=600): Figure'ü diske kaydet.
# .pdf: Vektör formatı. Grafik yeniden boyutlansa da kalitesini kaybetmez (genellikle tercih edilir).
# dpi=600: Kaydedilen dosyanın çözünürlüğünü 600 Nokta/İnç (DPI) olarak ayarla.
# PDF vektörel olduğu için DPI değeri teknik olarak gerekmez, ancak
# Matplotlib yine de bu parametreyi destekler ve bitmap (PNG/JPG) için kritiktir.
fig.savefig("figure_yuksek_kalite.pdf", dpi=600)

# PNG formatında yüksek çözünürlükle kaydetme
# .png: Bitmap (piksel tabanlı) formatı. Yüksek kaliteli çıktı için yüksek DPI (örneğin 600) gerektirir.
fig.savefig("figure_yuksek_kalite.png", dpi=600)

# Oluşturulan Figure'ü (içindeki grafikle birlikte) ekranda göster.
plt.show()

# Kullanıcıya dosyaların başarıyla kaydedildiğini bildir.
print("Yüksek çözünürlüklü grafikler 'figure_yuksek_kalite.pdf' ve '.png' adıyla kaydedildi.")
```
