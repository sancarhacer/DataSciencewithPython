```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
age_list = [10,20,30,30,30,40,50,60,70,75]
weight_list = [20,60,80,85,86,87,70,90,95,90]
```

```python
plt.plot(age_list,weight_list,"r")
plt.xlabel("age")
plt.ylabel("weight")
plt.title("""age vs weight""")

plt.show()
```

```python
#works with numpy
```

```python
np_age_list = np.array(age_list)
np_weight_list = np.array(weight_list)
np_age_list
```

```python
plt.plot(np_age_list,np_weight_list,"g")
plt.xlabel("age")
plt.ylabel("weight")
plt.title("""age vs weight""")

plt.show()
```

```python
numpy_arr1 = np.linspace(0,10,20)
numpy_arr1
numpy_arr2 = numpy_arr1 **2
```

```python
plt.plot(numpy_arr1,numpy_arr2,"r*-")
plt.show()
```

```python
plt.subplot(1,2,1)
plt.plot(numpy_arr1,numpy_arr2,"r*-")
plt.subplot(1,2,2)
plt.plot(numpy_arr2,numpy_arr1,"g*-")
plt.show()
```


##customize the axes

```python
# ----------------- FIGURE VE AXES OLUŞTURMA -----------------

# my_figure = plt.figure(): Yeni, boş bir Figure (ana tuval/pencere) nesnesi oluşturur.
my_figure = plt.figure()

# figure_axes = my_figure.add_axes([sol, alt, genişlik, yükseklik]):
# Figure üzerine manuel olarak bir Axes (alt grafik alanı) ekler.
# [0.1, 0.1, 0.5, 0.3]: Bu dört değer, Axes'in Figure içindeki konumunu ve boyutunu
# Figure'ün normalize edilmiş koordinatlarına (0'dan 1'e kadar) göre belirler.
# - sol: 0.1 (Figure'ün sol kenarından %10 içeride başla)
# - alt: 0.1 (Figure'ün alt kenarından %10 yukarıda başla)
# - genişlik: 0.5 (Figure'ün genişliğinin %50'si kadar genişlik)
# - yükseklik: 0.3 (Figure'ün yüksekliğinin %30'u kadar yükseklik)
figure_axes = my_figure.add_axes([0.1, 0.1, 0.5, 0.3])

# ----------------- ÇİZİM VE ETİKETLEME -----------------

figure_axes.plot(numpy_arr1, numpy_arr2, "g")

# X ekseni etiketini ayarla.
figure_axes.set_xlabel("X Label")

# Y ekseni etiketini ayarla.
figure_axes.set_ylabel("Y Label")

# Alt grafiğin başlığını ayarla.
figure_axes.set_title("Graph Title")

# ----------------- GÖSTERİM -----------------

# plt.show(): Oluşturulan Figure'ü ve içindeki Axes'i ekranda göster.
plt.show()
```

##customize labels

```python
new_fig = plt.figure(dpi=50)
# dpi=50: Figure'ün başlangıç çözünürlüğünü 50 Nokta/İnç (DPI) olarak ayarlar.
# Not: Bu başlangıç DPI değeri, ekran gösterimi için kullanılır, ancak fig.savefig() fonksiyonu ile kaydederken ayarlanan
# DPI değeri, dosyanın nihai baskı/kayıt çözünürlüğünü belirlemesi açısından daha önemlidir.
new_axes = new_fig.add_axes([0.1,0.1,0.5,0.9])
new_axes.plot(numpy_arr1,numpy_arr1 ** 2,label = "numpy array ** 2")
new_axes.plot(numpy_arr1, numpy_arr1 ** 3, label= "numpy array **3 ")
new_axes.legend(loc=6)
plt.show()
```

```python
# Aynı Axes Üzerinde Çoklu Çizim
# ax.plot() fonksiyonuna ardışık X Y çiftleri vererek aynı grafikte birden fazla çizgi çizebilirsiniz.
data1 = np.linspace(0,10,20)
data2 = data1 ** 2
my_fig, my_axes = plt.subplots()
my_axes.plot(data1, data2, color="#3A95A8",alpha=0.3)
my_axes.plot(data2, data1, color="#C96F23")
plt.show()
```

```python
(new_fig, new_axes) = plt.subplots()
new_axes.plot(data1, data1 + 2, color="red", linewidth = 1.0)
new_axes.plot(data1, data1 + 3, color ="yellow", linewidth = 3.0)

new_axes.plot(data1, data1 + 4, color = "#C96F23", linestyle = "-.")
new_axes.plot(data1, data1 + 5, color = "#C96F23", linestyle = ":")
new_axes.plot(data1, data1 + 6, color = "#C96F23", linestyle = "--")

new_axes.plot(data1, data1 + 7, color = "#000000", linestyle = "--", marker = "o",markersize = 8, markerfacecolor="blue")
new_axes.plot(data1, data1 + 8, color = "#000000", linestyle = "--", marker = "+",markersize = 4)
```

```python
(new_fig, new_axes) = plt.subplots(2,3)
new_axes[0,0].plot(data1, data1 , color="blue", linewidth = 1.0)
new_axes[0,1].plot(data1, data1**2 , color="red", linewidth = 3.0)
new_axes[0,2].plot(data2,data1,color = "#C96F23", linestyle = "-.")
new_axes[1,0].plot(data1, data1**2 /2 , color = "#000000", linestyle = "--", marker = "o",markersize = 8, markerfacecolor="red")
new_axes[1,1].plot(data1, data1**2 , color="red", linewidth = 1.0)
new_axes[1,2].plot(data2,data1,color = "#C96F23", linestyle = "-.")
```
