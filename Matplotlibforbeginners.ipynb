```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

#Grafiğini çizmek istediğiniz veriyi oluşturursunuz. Matplotlib genellikle NumPy dizileriyle çalışır
# .Örnek: $\sin(x)$ eğrisini çizmek için $0$ ile $10\pi$ arasında 1000 nokta oluşturulur.

# X ekseni verisini (0'dan 10*pi'ye kadar 1000 eşit aralıklı nokta) oluştur
X = np.linspace(0, 10 * np.pi, 1000)
# Y ekseni verisini (X'in sinüsü) oluştur
Y = np.sin(X)
```

plt.subplots(): Bir Figure (ana pencere/tuval) ve bir Axes (gerçek çizimin yapıldığı alt grafik) oluşturmanın en yaygın ve önerilen yoludur. Genellikle fig ve ax olarak adlandırılırlar.

ax.plot(X, Y): $\mathbf{X}$ ve $\mathbf{Y}$ verilerini kullanarak $\mathbf{ax}$ üzerinde bir çizgi grafiği çizer.

plt.show(): Grafiği ekranda gösterir.

```python

fig, ax = plt.subplots() # Figure (fig) ve Axes (ax) oluştur
ax.plot(X, Y)           # ax üzerinde çizgi grafiği çiz
plt.show()              # Grafiği göster
```

Matplotlib, veri setinize uygun çeşitli grafik türleri sunar.

```python
fig, ax = plt.subplots(figsize=(6, 4)) # Tek bir örnek için fig, ax oluştur
# Yeni bir boş çizim penceresi (fig) oluşturur ve boyutunu  6 x 4 inç olarak ayarlar.Bu pencerenin içine tek bir alt grafik alanı (ax) yerleştirir.

# 1. Scatter Plot (Serpilme Grafiği)
X_scatter = np.random.uniform(0, 1, 100) # 0 ile 1 arasında 100 rastgele X değeri
Y_scatter = np.random.uniform(0, 1, 100) # 0 ile 1 arasında 100 rastgele Y değeri
ax.scatter(X_scatter, Y_scatter)
ax.set_title("Serpilme Grafiği (ax.scatter)")
plt.show() # Her bir örneği ayrı ayrı göstermek için yorum satırlarını kaldırın
```

```python
fig, ax = plt.subplots(figsize=(6, 4)) # Tek bir örnek için fig, ax oluştur

# 2. Bar Plot (Çubuk Grafiği)
X_bar = np.arange(10)                       # 0'dan 9'a kadar tamsayılar
Y_bar = np.random.uniform(1, 10, 10)        # 1 ile 10 arasında 10 rastgele yükseklik
ax.bar(X_bar, Y_bar)
ax.set_title("Çubuk Grafiği (ax.bar)")
plt.show()
```

```python
# 3. Histogram
fig, ax = plt.subplots(figsize=(6, 4)) # Tek bir örnek için fig, ax oluştur

Z_hist = np.random.normal(0, 1, 1000) # Standart normal dağılımdan 1000 rastgele değer
ax.hist(Z_hist, bins=20) # 20 kutu ile histogram çiz
ax.set_title("Histogram (ax.hist)")
plt.show()
```

```python
# 4. Boxplot (Kutu Grafiği)
fig, ax = plt.subplots(figsize=(6, 4)) # Tek bir örnek için fig, ax oluştur
Z_box = np.random.normal(0, 1, (100, 3)) # 3 veri serisi için 100'er veri noktası
ax.boxplot(Z_box)
ax.set_title("Kutu Grafiği (ax.boxplot)")
plt.show()
```

Matplotlib, çizimlerin görünümünü değiştirmek için sınırsız esneklik sunar. Çizgi rengini, stilini, kalınlığını ve işaretleyicilerini (marker) değiştirebilirsiniz.

```python
# Grafikte kullanılacak veriyi oluştur
X = np.linspace(0, 10, 100) # 0 ile 10 arasında eşit aralıklı 100 adet X noktası oluştur
Y = np.sin(X)               # X noktalarının sinüs değerlerini Y olarak hesapla

# Bir Figure (fig) ve 4 adet alt Axes (grafik alanı) oluştur.
# (4, 1) demek: 4 satır, 1 sütun (yani alt alta 4 grafik).
# figsize=(8, 8) ile tüm Figure'ün boyutunu 8x8 inç olarak ayarla.
fig, axes = plt.subplots(4, 1, figsize=(8, 8))

# ----------------- 1. Renk (color) Özelliği -----------------
# axes[0]: Oluşturulan 4 Axes'in ilkine (indeks 0) eriş.
# color="red": Çizginin rengini kırmızı yap.
axes[0].plot(X, Y, color="red")
# Bu alt grafiğe bir başlık ekle.
axes[0].set_title("color='red'")

# ----------------- 2. Çizgi Stili (linestyle) Özelliği -----------------
# axes[1]: İkinci alt grafiğe (indeks 1) eriş.
# linestyle="--": Çizgi stilini kesikli çizgi ("dashed") olarak ayarla.
axes[1].plot(X, Y, linestyle="--")
# Bu alt grafiğe bir başlık ekle.
axes[1].set_title("linestyle='--'")

# ----------------- 3. Çizgi Kalınlığı (linewidth) Özelliği -----------------
# axes[2]: Üçüncü alt grafiğe (indeks 2) eriş.
# linewidth=5: Çizginin kalınlığını 5 birime ayarla (varsayılan 1 civarındadır).
axes[2].plot(X, Y, linewidth=5)
# Bu alt grafiğe bir başlık ekle.
axes[2].set_title("linewidth=5")

# ----------------- 4. İşaretleyici (marker) Özelliği -----------------
# axes[3]: Dördüncü alt grafiğe (indeks 3) eriş.
# marker="o": Veri noktalarına daire ("o") şeklinde işaretleyiciler ekle.
# markevery=10: Sadece her 10. veri noktasına işaretleyici çiz (grafiği daha temiz tutar).
axes[3].plot(X, Y, marker="o", markevery=10)
# Bu alt grafiğe bir başlık ekle.
axes[3].set_title("marker='o'")

# Alt grafikleri otomatik olarak düzenle. Başlıkların ve eksen etiketlerinin
# birbirine çakışmamasını sağlar.
plt.tight_layout()

# Oluşturulan Figure'ü (içindeki 4 alt grafikle birlikte) ekranda göster.
plt.show()
```

Organizasyon (Organize):

Bir şekilde birden fazla veri seti çizilebilir veya bir Figure'ü birden fazla alt parçaya (Axes) bölebilirsiniz

Aynı Axes Üzerinde Çoklu Çizim

ax.plot() fonksiyonuna ardışık $X, Y$ çiftleri vererek aynı grafikte birden fazla çizgi çizebilirsiniz.

```python
X = np.linspace(0, 10, 100)
Y1, Y2 = np.sin(X), np.cos(X)

fig, ax = plt.subplots()
# Y1 ve Y2'yi aynı Axes üzerinde çiz
ax.plot(X, Y1, label="Sinüs")
ax.plot(X, Y2, label="Kosinüs")
ax.legend() # Etiketleri göstermek için lejant ekle
ax.set_title("Aynı Grafikte İki Çizgi")
plt.show()
```

Subplots (Alt Grafikler)

plt.subplots(satır_sayısı, sütun_sayısı) kullanarak grafiği alt parçalara ayırabilirsiniz.

```python
# 2 satır, 1 sütun (yan yana)
fig, (ax1, ax2) = plt.subplots(2, 1, figsize=(8, 4))

ax1.plot(X, Y1, color="C0") # C0 varsayılan ilk renktir (genellikle mavi)
ax1.set_title("Sinüs Eğrisi")

ax2.plot(X, Y2, color="C1") # C1 varsayılan ikinci renktir (genellikle turuncu)
ax2.set_title("Kosinüs Eğrisi")

plt.tight_layout()
plt.show()
```

Etiketleme (Label)
Grafiğinizin anlaşılması için başlıklar ve eksen etiketleri eklemek kritiktir.

fig.suptitle(str): Figure'ün (tüm pencereni) ana başlığı.

ax.set_title(str): Alt grafiğin (Axes'in) başlığı.

ax.set_xlabel(str): X ekseni etiketi.

ax.set_ylabel(str): Y ekseni etiketi.

```python

fig, ax = plt.subplots()
ax.plot(X, Y)

# Etiketleme
fig.suptitle("Ana Grafik Başlığı") # En üstteki başlık
ax.set_title("Sinüs Dalgası")      # Alt grafik başlığı
ax.set_xlabel("Zaman (saniye)")    # X eksenini etiketle
ax.set_ylabel("Genlik (Birim)")    # Y eksenini etiketle

plt.show()
```

Kaydetme (Save)

Grafiğinizi bir görüntü dosyası olarak kaydetmek çok önemlidir. Vektör (örneğin .pdf) veya bitmap (örneğin .png) formatlarında kaydedebilirsiniz. Vektör formatları kaliteden ödün vermez ve baskı için idealdir.

```python
# fig, ax zaten yukarıdaki örnekten mevcut olduğunu varsayalım

# PNG formatında, 300 dpi (yüksek çözünürlük) ile kaydet
fig.savefig("my-first-figure.png", dpi=300)

# PDF formatında (Vektör formatı) kaydet
fig.savefig("my-first-figure.pdf")

print("Grafikler Colab'ın dosya sistemine kaydedildi.")
```
