```python
import pandas as pd
df = pd.read_csv("8-apply_function_data.csv")
df
```

```python
df.columns
```

```python
df[1:20]
```

```python
# Example 1: Apply a function to classify employees based on salary
def salary_category(salary):
    if salary < 50000:
        return "Low"
    elif 50000 <= salary < 80000:
        return "Medium"
    else:
        return "High"
df["Salary_Category"] = df["Salary"].apply(salary_category)
df
```

```python
# Example 2: Apply a function to calculate adjusted performance score
def adjust_performance(row):
    if row["Experience"] > 10:
        return row["Performance_Score"] + 1
    return row["Performance_Score"]
df["Adjusted_Performance"] = df.apply(adjust_performance, axis=1)
df
```

```python
# Example 3: Apply a lambda function to format employee names
df["Name"] = df["Name"].apply(lambda x: x.replace("_", " "))
df.head()
```
